```groovy
import static com.twosigma.beaker.widgets.DisplayAnyWidget.display;
import static com.twosigma.beaker.groovy.widgets.Interactive.interact;
import static com.twosigma.beaker.SerializeToString.showInternalWidget;

class PlotHolder{
    static String text = ""
    static Plot plot = null;
}

public static Object f(x){
    PlotHolder.text = x;
    if(PlotHolder.plot != null){
        PlotHolder.plot.close();
    }
    PlotHolder.plot = new Plot(title: PlotHolder.text)
    display(PlotHolder.plot) //TODO move to DisplayWidget.display ?
    PlotHolder.text
}

interact(this.&f, "s");
```
